```python
import os
from getpass import getpass

from agents import Agent, Runner

MODEL = "gpt-5.4-mini"
```

```python
if not os.environ.get("OPENAI_API_KEY"):
    os.environ["OPENAI_API_KEY"] = getpass("Enter your OpenAI API key: ")
```

```python
agent = Agent(
    name="Coding Assistant",
    instructions="You are a helpful coding assistant. Answer programming questions clearly and concisely.",
    model=MODEL,
)
```

```python
# Run a single query
result = await Runner.run(agent, "What are the SOLID principles in software engineering? Be brief.")
print(result.final_output)
```

```python
# Run another query — each call is stateless (no memory of previous)
result = await Runner.run(agent, "Write a Python function to check if a number is prime.")
print(result.final_output)
```

## Summary

- Created a basic agent with the OpenAI Agents SDK using `Agent()` and `Runner.run_sync()`.
- The agent has no tools — it can only generate text responses.
- Each `Runner.run_sync()` call is stateless; there is no conversation memory between calls.
